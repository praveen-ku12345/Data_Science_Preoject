```python
import pandas as pd
# Load the dataset
file_path = "C:\\Users\\prave\\Downloads\\creditcard.csv\\creditcard.csv"
df = pd.read_csv(file_path)
# Display basic information about the dataset
df.info(), df.head()
```

```python
from sklearn.preprocessing import StandardScaler
# Normalize the 'Amount' and 'Time' columns
scaler = StandardScaler()
df['Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
df['Time'] = scaler.fit_transform(df['Time'].values.reshape(-1, 1))
# Display the first few rows to confirm normalization
df[['Time', 'Amount']].head()
```

```python
from collections import Counter
from imblearn.over_sampling import SMOTE
# Check the class distribution
class_counts = Counter(df['Class'])
print(class_counts)
# Separate features and target variable
X = df.drop('Class', axis=1)
y = df['Class']
# Apply SMOTE to balance the classes
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
# Check the class distribution after resampling
print(Counter(y_resampled))
```

```python
from sklearn.model_selection import train_test_split
# Split the resampled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
# Initialize the models
rf_model = RandomForestClassifier(random_state=42)
lr_model = LogisticRegression(max_iter=1000, random_state=42)
# Train the models
rf_model.fit(X_train, y_train)
lr_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
# Predict using the models
rf_predictions = rf_model.predict(X_test)
lr_predictions = lr_model.predict(X_test)
# Evaluate the models
rf_report = classification_report(y_test, rf_predictions)
lr_report = classification_report(y_test, lr_predictions)
print("Random Forest Classification Report:\n", rf_report)
print("Logistic Regression Classification Report:\n", lr_report)
```

Use Random Forest for our credit card fraud detection project, as it provides better performance in identifying fraudulent transactions while maintaining a good balance between precision and recall
